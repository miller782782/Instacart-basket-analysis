### 5. Data consistency checks

```python
cust.describe()
```

```python
# Check for mixed-type data

for col in cust.columns.tolist():
  weird = (cust[[col]].map(type) != cust[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (cust[weird]) > 0:
    print (col)
```
