## Grouping Data with pandas

```python
df.groupby('product_name')
```

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
df.groupby('department_id')['order_number'].mean()
```

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
ords_prods_merge_2['max_order'] = ords_prods_merge_2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge_2.head(15)
```

```python
pd.options.display.max_rows = None
```

```python
ords_prods_merge_2.loc[ords_prods_merge_2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge_2.loc[(ords_prods_merge_2['max_order'] <= 40) & (ords_prods_merge_2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge_2.loc[ords_prods_merge_2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
