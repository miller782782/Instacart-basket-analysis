### 2. Find aggregated mean of 'order_number' grouped by 'department_id'

```python
#  Aggregate mean order_number grouped by department_id

ords_prods_merge_2.groupby('department_id').agg({'order_number': ['mean']})
```

### 4. Creating loyalty flags for existing customers

```python
# Group by 'user_id' and add a column containing the max number of orders.

ords_prods_merge_2['max_order'] = ords_prods_merge_2.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge_2.head(15)
```

```python
# Add the loyalty_flag column and classify each customer as Loyal, Regular or New

ords_prods_merge_2.loc[ords_prods_merge_2['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge_2.loc[(ords_prods_merge_2['max_order'] <= 40) & (ords_prods_merge_2['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge_2.loc[ords_prods_merge_2['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge_2.head(15)
```

```python
ords_prods_merge_2['loyalty_flag'].value_counts(dropna = False)
```

```python
# The loyalty flags have been successfully added
```

### 5. Check basic statistics for each loyalty category

```python
# Look at the mean, median, min and max prices for items purchased by each loyalty group of customers

ords_prods_merge_2.groupby('loyalty_flag').agg({'prices': ['mean', 'median', 'min', 'max']})
```

##### It looks as if new customers are attracted to the more expensive items averaging 13.29 per item whereas regular customers spend on average 0.79 less per item averaging 12.50.  Meanwhile the loyal customers average considerably less at just 10.39 per item.  The max price seems unrealistic and will require further investigation.

### 6. Create a spending flag for each user based on the average price across all their orders

```python
#  First let's find the average item price for each user

ords_prods_merge_2.groupby('user_id').agg({'prices': ['mean', 'median', 'min', 'max']})
```

```python
#  Now that we can see some of these figures, we can add a column for the 'mean_item_price'

ords_prods_merge_2['mean_item_price'] = ords_prods_merge_2.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge_2.head()
```

```python
# Add the spending_flag column and classify each customer as High spender or Low spender

ords_prods_merge_2.loc[ords_prods_merge_2['mean_item_price'] < 10, 'spender_flag'] = 'Low spender'
```

```python
ords_prods_merge_2.loc[ords_prods_merge_2['mean_item_price'] >= 10, 'spender_flag'] = 'High spender'
```

```python
ords_prods_merge_2.head()
```

```python
ords_prods_merge_2['spender_flag'].value_counts(dropna = False)
```

### 7. Creating a flag to distinguish frequent vs non-frequent customers

```python
# Create a column median_days_since_prior_order

ords_prods_merge_2['median_days_since_prior_order'] = ords_prods_merge_2.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge_2.head()
```

```python
# Now add a frequent customer flag classiying each customer as "Non-frequent customer", 
# "Regular customer" or "Frequent customer".

ords_prods_merge_2.loc[ords_prods_merge_2['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge_2.loc[(ords_prods_merge_2['median_days_since_prior_order'] > 10) & (ords_prods_merge_2['median_days_since_prior_order'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge_2.loc[ords_prods_merge_2['median_days_since_prior_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```
